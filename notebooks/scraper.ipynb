```python
from selenium import webdriver
import pandas as pd
import time
```

```python
driver = webdriver.Safari()
driver.get("https://www.rtings.com/monitor/tools/table/104496")

time.sleep(5)

scroll_div = driver.find_element("xpath", "/html/body/div/div[3]/div/div/div/div[2]/div[1]/div/div");


for i in range(60):
    driver.execute_script("arguments[0].scrollTo(0,20000)", scroll_div);
    i+=1;

table = driver.find_element("xpath", "/html/body/div/div[3]/div/div/div/div[2]/div[1]/div/div/table");

otable = table.get_attribute("outerHTML");
df = pd.read_html(otable)[0];
```
